# 1. Kurze Zusammenfassung: freies Elektronengas  <a id='freieselektronengas'></a>

## Annahmen

1. Keine Wechselwirkung zwischen Elektronen und Atomrümpfen
2. Keine Wechselwirkung zwischen Elektronen
3. Elekronen sollen sich in einem 3D-Kasten mit unendlich hohen unendlich hohen Potentialwänden und Volumen $V=L_xL_yL_z$ aufhalten

**Stationäre Schrödinger Gleichung (SGL) für ein Elektron**
\begin{equation}
-\frac{\hbar^2}{2m_e}\nabla^2\psi(\mathbf{r}) = E\psi(\mathbf{r})
\end{equation}

was der SGL für ein Teilchen im 3D-Kasten entspricht. Wir verwenden nun noch periodische Randbedingungen. 

Im Modell des freien Elektronengases betrachten wir nicht wechselwirkende Elektronen, die in einem 3-dimensional Kasten mit Volumen $V=L_xL_yL_z$ von unendlich hohen Potentialwänden eingeschlossen sind. Es genügt die Schrödinger Gleichung für ein einzelnes Elektron zu lösen, da die Elektronen, gemäß unseren Annahmen, nicht wechselwirken. Das ist das Standardproblem des "Teilchens im Kasten".  Wir führen nun zusätzlich noch periodische Randbedingungen ein. 

## Born–von Karman periodische Randbedingungen
<div class="alert alert-block alert-info"><b></b>
\begin{equation}
\Psi_{\mathbf{k}}(x,y,z)=\Psi_{\mathbf{k}}(x+L_x,y,z) =\Psi_{\mathbf{k}}(x,y+L_y,z)=\Psi_{\mathbf{k}}(x,y,z+L_z)
\end{equation}
</div>
&nbsp;

<center><img src="images/pbc.png" width="300" /></center>

Das einfachste Randbedingungen sind, dass die Wellenfunktion an den Rändern verschwinden muss (wie beim Teilchen im 3D Kasten). Dies entspricht dem Festhalten der Atome am Ende einer endlichen Kette. Damit erhält man stehende Wellen. Es ist unpraktisch, hier solche Randbedingungen zu verwenden, weil wir es letztendlich nicht-stationäre Wellen (Gruppengeschwindigkeit ungleich null) zur Berücksichtigung der elektrischen und thermischen Leitfähigkeit brauchen.Daher sind die periodischen Randbedingungen die bessere Wahl.

## Lösung für das freie Elektronengas

 <div class="alert alert-block alert-success"><b>Eigenfunktionen sind ebene Wellen:</b>
\begin{equation}
   \Psi_{\mathbf{k}}(\mathbf{r}) = \frac{1}{\sqrt{V}}\exp^{i\mathbf{k}\cdot\mathbf{r}} \quad \mathrm{mit\,Wellenvektor\,}\mathbf{k}=(k_x,k_yk_z) 
  \end{equation}
</div>


<div class="row">
  <div class="col-md-3" markdown="1">
   <div class="alert alert-block alert-success"><b>Energieeigenwerte</b>
  \begin{equation}
    E(\mathbf{k}) = \frac{\hbar^2k^2}{2m_e} 
   \end{equation}
</div>
  </div>
  <div class="col-md-1" markdown="1">
  </div>
  <div class="col-md-8" markdown="1">
    <div class="alert alert-block alert-success"><b>Zulässige Wellenzahlen</b>
   \begin{equation}
   k_x =\frac{2\pi}{L_x} n_x, \quad k_y =\frac{2\pi}{L_y} n_y, \quad k_z =\frac{2\pi}{L_z} n_z \quad \mathrm{mit} \quad n_{x,y,z}\in\mathbb{Z}
   \end{equation}
  </div>
  </div>
</div>

Die ebenen Elektronenwellen $\Psi_{\mathbf{k}}(\mathbf{r})$ haben die de Broglie Wellenlänge $\lambda = 2\pi/|\mathbf{k}|$ und die Normierung

$$
\int d^3r |\Psi_{\mathbf{k}}(\mathbf{r})|^2 = 1
$$

Die Wahrscheinlichkeit das Elektron, dass sich das Elektron irgendwo im Kasten befindet ist also 1. 

Einsetzen der zulässigen Wellenzahlen gibt:

\begin{equation}
E(\mathbf{k}) = \frac{\hbar^2}{2m_e}\left[k_x^2 + k_y^2 + k_z^2\right] = \frac{\hbar^2}{2m_e}\left[\frac{(2\pi)^2}{L_x^2}n_x^2 + \frac{(2\pi)^2}{L_y^2}n_y^2 + \frac{(2\pi)^2}{L_z^2}n_z^2\right]
\end{equation}

```python
from numpy import linspace 
import matplotlib.pyplot as plt 

nx = linspace(-14,14,30) # create a float array with values from 0.1 - 10 with 100 intervals
E = nx**2

plt.figure(figsize=(7,7)) # set the figsize
plt.rc("font", size=20)
plt.xlim(-15,15)
plt.ylim(-5,200)

plt.xlabel("$n_x$" "\n" "$k_x/(2\pi/L_x)$")
plt.ylabel("Energy $(h^2/L_x^2m_e)$") 
plt.plot(nx,E,'o--')

#Add text to plot
text_position = (0, 175)  
text_string = '$n_y=n_z=0$'
plt.text(*text_position, text_string)
```

## Grenzen/offene Fragen

- Gute Beschreibung von Vorhersage von Eigenschaften von Alkalimetallen, aber fuer 3$d$-Übergangsmetallen teils grosse Abweichungen vom Experiment 
- Warum sind manchne Festkörper Metalle und andere Halbleiter/Isolatoren? 
- Warum werden positive Werte fuer die Hall-Konstante gemessen
- ...

Der experimentelle Wert für die (elektronische) spezifische Wärmekapazität stimmt fuer Alkalimetalle gut mit dem Vorhersage vom freien Elektronengas überein, während es für 3$d$-Übergangsmetalle grosse Abweichungen zwischen Theorie und Experiment gibt (siehe Gross-Marx, Kapitel 7). Für 3$d$ Materialien sind die 3$d$-Elektronen, die auch massgeblich zur Zustandsdichte am Fermi-Niveau beitragen, stark lokalisiert und können nicht gut mit einem Modell von delokalisieren freien Elektronen beschrieben werden. 

Das Modell des freien Elektronengas beschreibt Metalle (mehr oder weniger) gut, aber es kann nicht erklären, warum Festkörper wie Silizium oder Diamant Halbleiter bzw. Isolatoren sind. Die äussere Elektronenschale von Kohlenstoff und Silizum ist nur halb gefüllt und man würde frei bewegliche Valenzelektronen erwarten. 


# 2. Bloch-Wellen: Elektronen im periodischen Potential <a id='periodicpot'></a>

 Erweiterung des Modell des freien Elektronengases: 
 
 Berücksichtigung des **periodischen Kristallpotentials** $U(\mathbf{r})$ der positiven Atomrümpfe

<div class="row">
  <div class="col-md-7" markdown="1">
    &nbsp;
      
<b>Annahmen:</b>
      
- Atomrümpfe sind starr
- nicht-wechselwirkende Elektronen 
  </div>
  <div class="col-md-5" markdown="1">
    <center><img src="images/blochpot.png" width="800" /></center>
  </div>
</div>

Im Fall des freien Elektronengases haben wir ein unendlich hohes Potential an den Rändern der Box mit Länge $L$, während das Potential konstant dazwischen ist. Bei der Erweiterung  beziehen wir ein, dass Atome (grüne Kugeln), Potentialtöpfe (Coulombpotential) erzeugen und, dass das Potential zwischen den Wändenn nicht flach ist. Das Bild zeigt eine 1D-Kette von Atomen, die mit dem Abstand $a$ separiert sind. 

In Festkörpern/Kristallen finden wir eine streng-periodische Anordnung on Atomen. Wir vernachlässigen Gitterschwingungen und gehen davon aus, dass die Atomrümpfe starr sind. Wir gehen weiterhin davon aus, dass die Elektronen nicht miteinander wechselwirken (Einelektronen-Näherung). Betrachtet man ein Elektron, so bewegt es sich im periodischen Potential $U(\mathbf{r})$, das von den Atomkernen hervorgerufen wird. Wir lösen die Schrödinger Gleichung für ein Elektron. Der $N-$Elektronenzustand wird erhalten, indem man die Einelektronenzustände unter der Berücksichtigung des Pauli-Prinzips auffüllt.

Wir betrachten die Lösungen der zeitunabhängigen Schrödinger Gleichung

\begin{equation}
H\Psi(\mathbf{r}) = \left[-\frac{\hbar^2}{2 m_e} + U(\mathbf{r})\right]\Psi(\mathbf{r}) = E\Psi(\mathbf{r})
\end{equation}

für ein einzelnes Elektron in Gegenwart eines gitterperiodisches Potentials 

\begin{equation}
U(\mathbf{r}) = U(\mathbf{r} + \mathbf{R})
\end{equation}

mit dem Translationsvektor $\mathbf{R}$ des dreidimensionalen Bravais-Gitters  

\begin{equation}
\mathbf{R} = n_1\mathbf{a}_1 + n_2\mathbf{a}_2 + n_3\mathbf{a}_3 \quad \mathrm{mit}\quad n_1,n_2,n_3\in \mathbb{Z}
\end{equation}

Die Vektoren $\mathbf{a}_1,\mathbf{a}_2, \mathbf{a}_3$ sind the primitiven Gittervektoren.


**Sind die elektronischen Zustände in einem Festkörper auch gitterperiodisch, d.h. $\psi(\mathbf{r}+\mathbf{R}) = \psi(\mathbf{r})$?**


<div class="row">
  <div class="col-md-10" markdown="1">
   <div class="alert alert-block alert-success"><b>Bloch-Theorem:</b>
      &nbsp;
        Die Eigenfunktionen $\Psi_{\mathbf{k}}(\mathbf{r})$ der Schrödinger Gl. sind gegeben als Produkt von ebenen Wellen mit einer gitterperiodischen Funktion
   
 \begin{equation}
   \Psi_{\mathbf{k}}(\mathbf{r}) = e^{i\mathbf{k}\cdot\mathbf{r}}u_{\mathbf{k}}(\mathbf{r}) \quad \mathrm{mit}
  \quad u_{\mathbf{k}}(\mathbf{r}) =   u_{\mathbf{k}} (\mathbf{r}+\mathbf{R})
 \end{equation}
    
  - $e^{i\mathbf{k}\cdot\mathbf{r}}$: ebene Welle
  - $u_{\mathbf{k}}(\mathbf{r})$: gitterperiodische Funktion        
</div>
  </div>
  <div class="col-md-2" markdown="1">
   <center><img src="images/felixbloch.png" width="240" /></center>
  </div>
</div>



<div class="alert alert-block alert-info"> <strong>Alternative Formulierung des Bloch Theorems</strong>    
\begin{equation}
   \Psi_{\mathbf{k}}(\mathbf{r}+\mathbf{R}) =  e^{i\mathbf{k}\cdot\mathbf{R}}\Psi_{\mathbf{k}}(\mathbf{r})
\end{equation}
</div>

<div class="alert alert-block-danger" role="alert">
    Die Eigenfunktion selbst sind also keine gitterperiodischen Funktionen,jedoch ihr Betragsquadrat.
</div>

Die alternative Form erhalten wir durch

\begin{equation}
\Psi_{\mathbf{k}}(\mathbf{r}+\mathbf{R}) = e^{i\mathbf{k}(\cdot\mathbf{r}+\mathbf{R})} u_{\mathbf{k}}(\mathbf{r}+\mathbf{R}) =  e^{i\mathbf{k}\cdot\mathbf{R}} e^{i\mathbf{k}\cdot\mathbf{r}}u_{\mathbf{k}}(\mathbf{r}) =  e^{i\mathbf{k}\cdot\mathbf{R}}\Psi_{\mathbf{k}}(\mathbf{r})
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 4 * np.pi, 1000)

# Calculate y values for u_k, Re[exp(ikr)], Psi and Psi²
#uk = np.sin(11*x)
uk = np.sin(6*x)+np.cos(6*2*x)+np.sin(6*3*x)
RePW = np.cos(x)
ImPW = np.sin(x)
RePsi = uk * RePW
ImPsi = uk* ImPW
Psi = RePsi + 1j * ImPsi
Psi2 = np.abs(Psi)**2

# Create subplots
fig, axs = plt.subplots(5, 1, figsize=(10, 10))
plt.rcParams['font.size'] = 18
plt.rcParams['font.family'] = "sans-serif"
for i in range(5):
    axs[i].set_xticklabels([])  # Remove x-labels
    axs[i].set_xticks([])
    axs[i].axhline(y=0,linestyle='--')
    axs[i].set_yticks([0])

# Plot u_k(x)
axs[0].plot(x, uk, color='olive')
axs[0].set_ylabel('$u_k(x)$')

# Plot Re(exp(ikx)) and Im(exp(ikx))
axs[1].plot(x, RePW, label='$Re(e^{ikx})$', color='red')
axs[1].plot(x, ImPW, label='$Im(e^{ikx})$', color='blue')
axs[1].set_ylabel('$e^{ikx}$')
axs[1].legend()

# Plot Re(Psi(x))
axs[2].plot(x, RePsi, color='red')
#axs[2].plot(x, ImPsi, label='Im()', color='blue')
axs[2].set_ylabel('$Re(\Psi_k(x))$')


# Plot Im(Psi(x))
axs[3].plot(x, ImPsi, color='blue')
axs[3].set_ylabel('$Im(\Psi_k(x))$')

# Plot Psi(x)^2
axs[4].plot(x, Psi2, color='green')
axs[4].set_ylabel('$\Psi_k(x)^2$')
axs[4].set_xlabel('$x$')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

## Periodische Randbedingungen

Wir betrachten periodische Randbedingungen auf zwei verschiedene Längenskalen:

1. **mikroskopisch** aufgrund der Symmmetry des Bravaisgitters gegeben durch $\mathbf{R}=\sum_i n_i \mathbf{a}_i$
2. **makroskopisch** aufgrund der Born-von Karman Randbedingungen: $\mathbf{L}$


### Werte des Wellenvektors

Wie beim freien Elektronengas können wir fordern, dass die Born-von Karman Randbedingungen gelten. Generalisiert auf nicht-kubische Systeme soll gelten

\begin{equation}
\Psi_{\mathbf{k}}(\mathbf{r})= \Psi_{\mathbf{k}}(\mathbf{r}+N_1\mathbf{a}_1)  = \Psi_{\mathbf{k}}(\mathbf{r}+N_2\mathbf{a}_2)   = \Psi_{\mathbf{k}}(\mathbf{r}+N_3\mathbf{a}_3)  
\end{equation}

wobei $N_1,N_2,N_3$ sehr grosse ganze Zahlen sind.Aufgrund des Bloch-Theorems muss gelten

\begin{equation}
\Psi_{\mathbf{k}}(\mathbf{r}+N_i\mathbf{a}_i) = e^{i\mathbf{k}\cdot(N_i\mathbf{a}_i)} \Psi_{\mathbf{k}}(\mathbf{r}) \stackrel{!}{=} \Psi_{\mathbf{k}}(\mathbf{r})
\end{equation}

Damit muss gelten:
\begin{equation}
e^{iN_1\mathbf{k}\mathbf{a}_1}= e^{iN_2\mathbf{k}\mathbf{a}_2} = e^{iN_3\mathbf{k}\mathbf{a}_3} = 1
\end{equation}

Dies kann nur gelten, wenn

\begin{equation}
\mathbf{k} = \frac{m_1}{N_1}\mathbf{b}_1 +  \frac{m_2}{N_2}\mathbf{b}_2 +  \frac{m_3}{N_3}\mathbf{b}_3 \quad \text{mit}\quad m \in \mathbb{Z}
\end{equation}
 
$N=N_1N_2N_3$ is the totale Anzahl primitiver Zellen im Kristall und $\mathbf{b}$ sind die reziproken Gittervektoren der Einheitszelle.  

$\mathbf{k}$ hat die Form

\begin{equation}
\mathbf{k} = x_1\mathbf{b}_1+x_2\mathbf{b}_2+x_3\mathbf{b}_3 \quad \mathrm{mit} \quad x_1,x_2,x_3\in\mathbb{Z}
\end{equation}

Es soll auch gelten $\mathbf{b}_i\mathbf{a}_j=2\pi\delta_{ij}$, wobei $\mathbf{a}_j$ die primitiven Gittervektoren im Realraum sind. Damit muss gelten

\begin{equation}
e^{i2\pi N_ix_i} = \cos(2\pi N_i x_i) + i \sin(2\pi N_i x_i) = 1
\end{equation}

Damit muss 

\begin{equation}
x_i = \frac{m_i}{N_i}
\end{equation}

sein.


Beim Elektronengas hat wir eine kubische Box angenommen und deshalb war $L_x=N_1|\mathbf{a}|_1$, $L_y=N_2|\mathbf{a}|_2$ und $L_z=N_3|\mathbf{a}|_3$

## Beweis Bloch-Theorem

Der Lösungsansatz ist wieder eine Linearkombination ebener Wellen

$$
\Psi(\mathbf{r}) = \sum_{\mathbf{k}}c_{\mathbf{k}}e^{i\mathbf{k}\cdot\mathbf{r}}
$$

wobei die Werte für $\mathbf{k}$ mit den Randbedingungen übereinstimmen und $c_{\mathbf{k}}$ die Normalisierungskonstante enthält.

Das gitterperiodische Potential kann auch als Fourier-Reihe geschrieben werden

$$
U(\mathbf{r}) = \sum_{\mathbf{G}} U_{\mathbf{G}}e^{i\mathbf{G}\cdot\mathbf{r}}
$$

wobei $\mathbf{G}$ ein reziproker Gittervektor ist

$$
\mathbf{G} = h\mathbf{b}_1+k\mathbf{b}_2+l\mathbf{b}_3 \quad \mathrm{mit} \quad h,k,l\in\mathbb{Z}
$$

Da das Potential eine reelle Zahl sein soll, gilt

$$
 U_{-\mathbf{G}} = U_{\mathbf{G}}^*
$$

Wir setzen nun den Lösungsansatz in den kinetischen Energieterm ein:

\begin{equation}
-\frac{\hbar^2\nabla^2}{2m_e}\Psi(\mathbf{r}) = \sum_{\mathbf{k}}\frac{\hbar^2k^2}{2m_e}c_{\mathbf{k}}e^{i\mathbf{k}\cdot\mathbf{r}}
\end{equation}

Für die potentielle Energie erhalten wir

\begin{align}
U(\mathbf{r})\Psi(\mathbf{r})=\left(\sum_{\mathbf{G}}U_{\mathbf{G}}e^{i\mathbf{G}\cdot\mathbf{r}} \right)\left(\sum_{\mathbf{k}}c_{\mathbf{k}}e^{i\mathbf{k}\cdot\mathbf{r}}\right)&=\left(\sum_{\mathbf{k}\mathbf{G}}U_{\mathbf{G}}c_{\mathbf{k}}e^{i(\mathbf{G}+\mathbf{k})\cdot\mathbf{r}}\right)\\&=\sum_{\mathbf{k}'\mathbf{G}}U_{\mathbf{G}}c_{\mathbf{k}'-\mathbf{G}}e^{i\mathbf{k}'\cdot\mathbf{r}}
\end{align}

wobei wir im letztem Schritt den Summationsindex umbenannt haben ($\mathbf{k} \rightarrow \mathbf{k}'=\mathbf{G}+\mathbf{k}$), um für die ebene Wellen den selben Ausdruck wie im kinetischen Energieterm zu haben. Die Verschiebung des k-indexes um den reziproken Gittervektor ist zulässig, da sich die Summe über alle Wellenvektoren erstreckt, die mit den periodischen Randbedingungen übereinstimmen, und die reziproken Gittervektoren sind eindeutig eine Teilmenge davon.

Wir benennen $\mathbf{k}'$ zurück zu $\mathbf{k}$ im Term für die potentielle Energie und können nun die Schrödinger Gleichung schreiben als

\begin{equation}
\sum_{\mathbf{k}}e^{i\mathbf{k}\cdot\mathbf{r}}\left[\left(\frac{\hbar^2k^2}{2m_e}-E\right)c_{\mathbf{k}}+\sum_{\mathbf{G}}U_{\mathbf{G}}c_{\mathbf{k}-\mathbf{G}}\right] = 0
\end{equation}


Die diese Gleichung für jeden Ortsvektor $\mathbf{r}$ erfüllt sein muss, muss der Ausdruck in den rechteckign Klammern, der unabhängig von $\mathbf{r}$ ist, für jeden Summand separat verschwinden. (Man kann auch mit Orthogonalitätsbedingungen argumentieren) 

\begin{equation}
\left(\frac{\hbar^2k^2}{2m_e}-E\right)c_{\mathbf{k}}+\sum_{\mathbf{G}}U_{\mathbf{G}}c_{\mathbf{k}-\mathbf{G}}= 0
\end{equation}



Erlaubte $\mathbf{k}$ Zustände im eindimensionalen $\mathbf{k}$ Raum:


<center><img src="images/kvalues.png" width="1200" /></center>

Wir legen nun fest, dass $\mathbf{k}$ in der 1. Brillouin-Zone liegen soll. In dieser Gleichung tauchen nur Koeffizienten $c_k$ auf, deren $\mathbf{k}$-Werte sich um einen reziproken Gittervektoren $\mathbf{G}$ unterscheiden (also nur ein kleiner Teil aller $c_k$), siehe Bild in vorheriger Box. Das heisst es werden die Koeffizienten $c_{\mathbf{k}}$, $c_{\mathbf{k}-\mathbf{G}'}$, $c_{\mathbf{k}-\mathbf{G}'}$,... gekoppelt. Damit zerfällt das Gleichungssystem in $N$ unabhängige Gleichungssysteme, i.e., in ein Gleichungssystem fuer jeden erlaubten $\mathbf{k}$-Vektor in der 1. Brillouin-Zone. Die Anzahl $N$ der möglichen Wellenvektoren in der 1. Brillouin-Zone wird gegeben durch (fuer den 1D-Fall) 

\begin{equation}
N=\frac{L_x}{a}
\end{equation}

i.e., $N$ ist die Einzahl von Einheitszellen im Kristall. 

Die Summe über $\mathbf{G}$ ist im Prinzip unendlich. In der Praxis findet man jedoch, dass typischerweise die Bestimmung von ein paar Koeffizienten ausreichend ist. 

\begin{equation}
\Psi_{\mathbf{k}}(\mathbf{r}) = \sum_{\mathbf{G}}c_{\mathbf{k}-\mathbf{G}}e^{i(\mathbf{k}-\mathbf{G})} =  e^{i\mathbf{k}\mathbf{r}}\left(\sum_{\mathbf{G}}c_{\mathbf{k}-\mathbf{G}}e^{-i\mathbf{G}\cdot\mathbf{r}}\right)
\end{equation}

Der Term in der Klammer ist eine Fourier Reiher ueber reziproke Gitterfunktionen und daher eine gitterperiodische Funktion, womit das Bloch-Theorem bewiesen wäre. 

## 3. Dispersionsrelation und Bandstruktur<a id='band'></a>

Beim Beweis des Bloch-Theorems hatten wir folgende Eigenwertgleichung stehen

\begin{equation}
\frac{\hbar^2k^2}{2m_e}c_{\mathbf{k}}+\sum_{\mathbf{G}}U_{\mathbf{G}}c_{\mathbf{k}-\mathbf{G}}= Ec_{\mathbf{k}}
\end{equation}

Wir wollen, dass $\mathbf{k}$ in der ersten Brillouin-Zone liegt. Wir können immer einen reziproken Gittervektor $\mathbf{G}$ finden, so dass der Wellenvektor $\mathbf{k}'+\mathbf{G}$ wieder in der 1. Brillouin-Zone liegt. 

\begin{equation}
\frac{\hbar^2|\mathbf{k}-\mathbf{G}|^2}{2m_e}c_{\mathbf{k}-\mathbf{G}}+\sum_{\mathbf{G'}}U_{\mathbf{G'}}c_{\mathbf{k}-\mathbf{G}-\mathbf{G}'}=  Ec_{\mathbf{k}-\mathbf{G}}
\end{equation}

wobei wir $\mathbf{k}'=\mathbf{k}-\mathbf{G}$ benutzt haben und nun die Umbenennung $\mathbf{G}''=\mathbf{G}'+\mathbf{G}$ vornehmen:

\begin{equation}
\frac{\hbar^2|\mathbf{k}-\mathbf{G}|^2}{2m_e}c_{\mathbf{k}-\mathbf{G}}+\sum_{\mathbf{G''}}U_{\mathbf{G''-\mathbf{G}}}c_{\mathbf{k}-\mathbf{G}''} = Ec_{\mathbf{k}-\mathbf{G}}
\end{equation}


Um das Eigenwertproblem zu lösen muessen wir foglende Matrix diagonlisieren
$$
H_{\mathbf{k}} =
\begin{bmatrix}
\frac{\hbar^2 |\mathbf{k}-\mathbf{G}_1|^2} {2m_e} + U_{\mathbf{G}_1-\mathbf{G}_1} & U_{\mathbf{G}_2-\mathbf{G}_1} & U_{\mathbf{G}_3-\mathbf{G}_1} &\cdots\\
 U_{\mathbf{G}_1-\mathbf{G}_2} & \frac{\hbar^2 |\mathbf{k}-\mathbf{G}_2|^2} {2m_e} + U_{\mathbf{G}_2-\mathbf{G}_2} & U_{\mathbf{G}_3-\mathbf{G}_2} & \dots \\
 \vdots & \vdots & \vdots &\ddots
\end{bmatrix}
$$
und erhalten die Eigenvektoren

$$
\mathbf{c}_{\mathbf{k}-\mathbf{G}} =
\begin{bmatrix}
c_{\mathbf{k}-\mathbf{G}_1} \\
c_{\mathbf{k}-\mathbf{G}_2}\\
c_{\mathbf{k}-\mathbf{G}_3}\\
\vdots
\end{bmatrix}
$$

und den Energieeigenwerten $E_n(\mathbf{k})$, wobei $n \in \mathbb{N}$ der sogenannte Bandindex ist.

<div class="alert alert-block alert-info">
  Wir erhalten fuer jeden erlaubten $\mathbf{k}$ Wert in der 1. Brilloiun-Zone einen Satz von Eigenenergien $E_n(\mathbf{k})$. 
</div>


<div class="alert alert-block alert-success">
   Bloch-Funktionen haben also noch den Bandindex $n$:   
   $$
   \Psi_{n\mathbf{k}}(\mathbf{r}) = e^{i\mathbf{k}\cdot\mathbf{r}}u_{n\mathbf{k}}(\mathbf{r})
   $$
  
</div>

## Qualitatives Bild der Energiebänder:


<center><img src="images/scheme_bandstructure.png" width="800" /></center>

Image source: "Festkörperphysik" by Rudolf Gross and Achim Marx, De Gruyter Studium, 2022

Die $\mathbf{k}$-Abhängigkeit der Energieeigenwerte $E_n$ wird als Dispersionsrelation bezeichnet. Die Schrödingergleichung liefert einen Satz von Eigenenergien $E_n(\mathbf{k})$, die mit dem Bandindex $n$ durchnummeriert sind. Obiges Bild zeigt ein qualitatives Bild der Energiebänder $E_n(\mathbf{k})$ als Funktion von $\mathbf{k}$.
